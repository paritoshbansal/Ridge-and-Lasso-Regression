```python
x=data.weight.values.reshape(-1,1)
y=data.mpg.values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lin=LinearRegression()
```

```python
model=lin.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,model.predict(x_test))
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
rd=Ridge()
```

```python
model2=rd.fit(x_train,y_train)
```

```python
print("training score=",model2.score(x_train,y_train))
```

```python
print("testing scoore=",r2_score(y_test,model2.predict(x_test)))
```

## Lasso Regression

```python
ls=Lasso()
```

```python
model3=ls.fit(x_train,y_train)
```

```python
print("training score=",model3.score(x_train,y_train))
```

```python
print("testing score=",r2_score(y_test,model3.predict(x_test)))
```
